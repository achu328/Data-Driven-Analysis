```python
import os
import yaml
import pandas as pd
```

```python
# 2023-11:

# Folder containing YAML files
input_folder = r"C:\Users\achu1\Documents\Project 2\data\2023-11"
output_folder = r"C:\Users\achu1\Documents\Project 2\output folder\2023-11"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Process each YAML file
for filename in os.listdir(input_folder):
    if filename.endswith(".yaml") or filename.endswith(".yml"):
        yaml_path = os.path.join(input_folder, filename)

        # Load YAML data
        with open(yaml_path, "r") as file:
            yaml_data = yaml.safe_load(file)

        if isinstance(yaml_data, list):
            df = pd.DataFrame(yaml_data)
        else:
            print("Next")

        # Save to CSV (same name as YAML file)
        csv_filename = os.path.splitext(filename)[0] + ".csv"
        csv_path = os.path.join(output_folder, csv_filename)
        df.to_csv(csv_path, index=False)
        print(f"Done")


## Concating all files:

# Folder containing the converted CSV files
csv_folder = r"C:\Users\achu1\Documents\Project 2\output folder\2023-11"
final_folder = r"C:\Users\achu1\Documents\Project 2\output folder\Merged CSV"

# List to store individual DataFrames
dfs = []

# Read each CSV file and append it to the list
for filename in os.listdir(csv_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(csv_folder, filename)
        df = pd.read_csv(file_path)
        dfs.append(df)

# Concatenate all DataFrames into one
merged_df1 = pd.concat(dfs, ignore_index=True)

# Save merged DataFrame to a new CSV (optional)
merged_df1.to_csv(os.path.join(final_folder, "merged_stocks1.csv"), index=False)

print("All CSV files have been merged successfully!")
```

## Cumulative Return Over Time:

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path =r"C:\Users\achu1\Documents\Project 2\output folder\Merged CSV\stock_df.csv"
df = pd.read_csv(file_path)


# Sort by Ticker and Date
df = df.sort_values(by=["Ticker", "date"])

# Calculate Daily Returns: (Close - Previous Close) / Previous Close
df["daily_return"] = df.groupby("Ticker")["close"].pct_change()

# Calculate Cumulative Return: (1 + daily return) cumulative product
df["cumulative_return"] = df.groupby("Ticker")["daily_return"].apply(lambda x: (1 + x).cumprod()).reset_index(level=0, drop=True) 

# Get the final cumulative return for each stock at year-end
final_cumulative_return = df.groupby("Ticker")[["date","cumulative_return"]].last().reset_index()

# Select the Top 5 Performing Stocks based on Cumulative Return
top_5_stocks = final_cumulative_return.nlargest(5, "cumulative_return")["Ticker"].tolist()

# Filter the DataFrame for only the Top 5 Stocks
top_5_df = df[df["Ticker"].isin(top_5_stocks)]

# Plot Cumulative Return for Top 5 Performing Stocks
plt.figure(figsize=(50,20))
for ticker in top_5_stocks:
    stock_data = top_5_df[top_5_df["Ticker"] == ticker]
    plt.plot(stock_data["date"], stock_data["cumulative_return"], label=ticker)

plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.title("Cumulative Return of Top 5 Performing Stocks Over the Year")
plt.legend()
plt.show()

# Print the Top 5 Performing Stocks
print("Top 5 Performing Stocks:\n", top_5_stocks)
```

## Top 5 Gainers and Losers(Month-wise): 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Convert "Date" column to datetime format
df_merged["date"] = pd.to_datetime(df_merged["date"], errors="coerce")

# Extract Year-Month for grouping
df_merged["Month"] = df_merged["date"].dt.to_period("M")

# Sort data by Date (important for calculating returns)
df = df_merged.sort_values(["Ticker", "date"])

# Calculate Monthly Returns
df["Previous_Close"] = df.groupby("Ticker")["close"].shift(1)  # Previous day's close
df["Monthly_Return"] = (df["close"] - df["Previous_Close"]) / df["Previous_Close"] * 100  # Percentage return

# Drop NaN values (first row of each stock will have NaN return)
df.dropna(subset=["Monthly_Return"], inplace=True)

# Group by Month and find top gainers & losers
top_gainers = df.groupby("Month").apply(lambda x: x.nlargest(5, "Monthly_Return")).reset_index(drop=True)
top_losers = df.groupby("Month").apply(lambda x: x.nsmallest(5, "Monthly_Return")).reset_index(drop=True)


# Combine both DataFrames
top_gainers["Category"] = "Top Gainers"
top_losers["Category"] = "Top Losers"

df_top_performers = pd.concat([top_gainers, top_losers]).reset_index(drop=True)

# # Display the output DataFrame
df_top_performers



# Plot Gainers & Losers for Each Month
months = df["Month"].unique()

for month in months:
    plt.figure(figsize=(12, 5))

    # Filter top gainers and losers for the specific month
    gainers = top_gainers[top_gainers["Month"] == month]
    losers = top_losers[top_losers["Month"] == month]

    # Plot Top 5 Gainers
    plt.subplot(1, 2, 1)
    sns.barplot(y=gainers["Ticker"], x=gainers["Monthly_Return"], palette="Greens_r")
    plt.title(f"Top 5 Gainers - (month)")
    plt.xlabel("Monthly Return (%)")
    plt.ylabel("Stock")

    # Plot Top 5 Losers
    plt.subplot(1, 2, 2)
    sns.barplot(y=losers["Ticker"], x=losers["Monthly_Return"], palette="Reds_r")
    plt.title(f"Top 5 Losers - (month)")
    plt.xlabel("Monthly Return (%)")
    plt.ylabel("Stock")

    plt.tight_layout()
    plt.show()
```
